```python
import keras
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
import scipy.ndimage

from PIL import Image
%matplotlib inline
```

```python
#############INPUT GENERATOR############

#Image folder
imageFolder = "./samples/"
images = os.listdir(imageFolder)


def tiff_to_npy(imageFolder):
    print("Converting tiffs...", end="")
    for x in sorted(os.listdir(imageFolder)):
        img = Image.open(imageFolder + x)
        try:
            img.seek(1)
            im = np.array(img)
            im = scipy.ndimage.zoom(im, 0.5, order=1)

        except EOFError:
            break
        flat = im    
        flat = (flat - np.min(flat))/(np.amax(im)- np.min(flat))

        yield flat

    print("Done")
```

```python
#############GENERATOR############
for i in tiff_to_npy(imageFolder):
    plt.imshow(i)
    plt.show()
    
```
